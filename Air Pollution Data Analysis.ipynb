### Time Lagged Cross Correlation

```python
def df_shifted(df, target=None, lag=0):
    if not lag and not target:
        return df       
    new = {}
    for c in df.columns:
        if c == target:
            new[c] = df[target]
        else:
            new[c] = df[c].shift(periods=lag)
    return  pd.DataFrame(data=new)
```

```python
df_from_2010 = df_low_sdi[df_low_sdi['year_id'] >= 2010]
df_from_2010
```

```python
df_new = df_shifted(df_from_2010[['mean_pm','prob_death_noshock']], 'prob_death_noshock', lag=-5)
df_new.corr()
```
