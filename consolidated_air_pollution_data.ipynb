### Creating a single Air Pollution dataframe

```python
air_pollution_df = pd.merge(ozone_df, pm_df,  how='outer', left_on=['location_id','year_id'], right_on = ['location_id','year_id'])
```

```python
air_pollution_df.head()
```

```python
air_pollution_df.columns
```

```python
air_pollution_df.drop(['measure_id_x', 'measure_id_y', 'measure_name_x', 'measure_name_y', 'location_name_y', 'rei_id_x', 'rei_id_y', 'rei_name_x', 'rei_name_y', 'unit_x', 'unit_y'], axis=1, inplace=True)
```

```python
air_pollution_df.head()
```

```python
air_pollution_df = pd.merge(air_pollution_df, hap_df[['location_id', 'year_id', 'mean_hap', 'median_hap', 'upper_hap', 'lower_hap', 'population_hap']],  how='outer', left_on=['location_id','year_id'], right_on = ['location_id','year_id'])
```

```python
air_pollution_df = pd.merge(air_pollution_df, loc_hier_df[['location_id', 'sdi_quintile', 'commonwealth', 'eu', 'four_world_regions', 'G20', 'WHO_regions', 'WB_income', 'WB_regions']],  how='left', left_on=['location_id'], right_on = ['location_id'])
```

### Combining with air pollution dataset

```python
combined_df = pd.merge(air_pollution_df, both_adult_mortality_noshock_df[['location_id', 'year_id', 'prob_death_noshock', 'upper_death_noshock', 'lower_death_noshock']],  how='left', left_on=['location_id', 'year_id'], right_on = ['location_id', 'year_id'])
```

```python
combined_df.shape
```

```python
combined_df = pd.merge(combined_df, both_adult_mortality_wshock_df[['location_id', 'year_id', 'prob_death_wshock', 'upper_death_wshock', 'lower_death_wshock']],  how='left', left_on=['location_id', 'year_id'], right_on = ['location_id', 'year_id'])
```

```python
combined_df.head()
```

```python
combined_df.shape
```

### Adding Air Pollution Score

```python
combined_df['mean_air_pollution'] = combined_df[['mean_ozone', 'mean_pm', 'mean_hap']].mean(axis=1)
```

```python
combined_df['sum_air_pollution'] = combined_df[['mean_ozone', 'mean_pm', 'mean_hap']].sum(axis=1)
```

```python
combined_df['sum_ozone_hap'] = combined_df[['mean_ozone', 'mean_hap']].mean(axis=1)
```

```python
combined_df.sum_air_pollution.quantile(0.33)
```

```python
combined_df.sum_air_pollution.quantile(0.66)
```

```python
combined_df['sum_air_pollution_cat'] = pd.cut(combined_df['sum_air_pollution'], bins=3, labels=['low', 'medium', 'high'])
```

```python
combined_df['sum_ozone_hap_cat'] = pd.cut(combined_df['sum_ozone_hap'], bins=3, labels=['low', 'medium', 'high'])
```
