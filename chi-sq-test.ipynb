### Chi-squared test to evaluate relationships between Air Pollution and SDI Quintile, WB Region and Income Countries

```python
crosstab = pd.crosstab(df.sum_ozone_hap_cat, df.sdi_quintile)
stats.chi2_contingency(crosstab)
```

```python
crosstab
```

```python
print('Test statistic:', stats.chi2_contingency(crosstab)[0])
print('p-value:', stats.chi2_contingency(crosstab)[1])
print('DOF:', stats.chi2_contingency(crosstab)[2])
```
